```python
from kafka import KafkaProducer

producer = KafkaProducer(bootstrap_servers="localhost:9092")
producer.send("test_topic", b"hello from jupyter")
producer.flush()
print("Sent from Jupyter")
```

```python
from kafka import KafkaProducer
producer = KafkaProducer(bootstrap_servers="localhost:9092")

for i in range(3):
    producer.send("test_topic", f"parquet test message {i}".encode("utf-8"))
producer.flush()

print("3 messages sent")
```
