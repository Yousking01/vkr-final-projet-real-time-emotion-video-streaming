```python
import json, time
from kafka import KafkaProducer

producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)

source_id = "cam_01"
events = [
    {"source_id": source_id, "emotion": "happy", "score": 0.93},
    {"source_id": source_id, "emotion": "neutral", "score": 0.72},
    {"source_id": source_id, "emotion": "sad", "score": 0.81},
    {"source_id": source_id, "emotion": "happy", "score": 0.95},
]

for e in events:
    e["event_time"] = time.strftime("%Y-%m-%d %H:%M:%S")
    producer.send("test_topic", e)
    time.sleep(0.3)

producer.flush()
print("✅ JSON events sent")
```

```python
import json, time, random
from kafka import KafkaProducer

producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)

def fake_bbox():
    x = random.randint(50, 200)
    y = random.randint(50, 200)
    w = random.randint(80, 140)
    h = random.randint(80, 140)
    return x, y, w, h

emotions = ["happy", "neutral", "sad", "angry", "surprise"]

for frame_id in range(1, 11):
    x, y, w, h = fake_bbox()
    event = {
        "source_id": "cam_01",
        "frame_id": frame_id,
        "face_id": 1,
        "producer_ts": time.time(),  # epoch seconds
        "emotion": random.choice(emotions),
        "score": round(random.uniform(0.55, 0.98), 2),
        "bbox_x": x, "bbox_y": y, "bbox_w": w, "bbox_h": h
    }
    producer.send("test_topic", event)
    time.sleep(0.2)

producer.flush()
print("✅ sent 10 pro events")
```

```python
import json, time, random
from kafka import KafkaProducer

producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)

emotions = ["happy", "neutral", "sad", "angry", "surprise"]

def fake_bbox():
    x = random.randint(40, 250)
    y = random.randint(40, 200)
    w = random.randint(80, 160)
    h = random.randint(80, 160)
    return x, y, w, h

for frame_id in range(1, 21):
    x, y, w, h = fake_bbox()
    event = {
        "source_id": "cam_01",
        "frame_id": frame_id,
        "face_id": 1,
        "producer_ts": time.time(),   # epoch seconds (float)
        "emotion": random.choice(emotions),
        "score": round(random.uniform(0.55, 0.98), 2),
        "bbox_x": x, "bbox_y": y, "bbox_w": w, "bbox_h": h
    }
    producer.send("test_topic", event)
    time.sleep(0.15)

producer.flush()
print("✅ sent 20 pro events")
```
